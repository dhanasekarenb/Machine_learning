```python
#List Comprehension

ex = [i**2 for i in range(5)]
print(ex)

ex_2 = [i ** 2 for i in range(5) if i % 2 == 0]
print(ex_2)

#matrix list
matrix = [[i for i in range(3)] for j in range(3)]
print(matrix)
```

```python
#Comprehension with mulitple if

option = {"ally", "alen", "world", "", "why"}

print([string
       for string in option
       if len(string) > 1
       if string[0] != 'a'
       ])
```

# _ _ call _ _ — Making an Object Callable

In PyTorch, nn.Module uses this to make model objects act like functions.

```python
class Adder:
  def __init__(self, x):
    self.x = x

  def __call__(self, y):
    return self.x + y

print(Adder(5)(6))

Add10 = Adder(10)
print(Add10(6))
```

# Inheritance

```python
class Base:
    def greet(self):
        print("Hello from Base")

class Derived(Base):
    def greet(self):
        super().greet()
        print("Hello from Derived")

obj = Derived()
obj.greet()
```

```python
#Challenge 3: Build a class LinearModel that:

#Takes w and b in __init__
#Implements __call__(x) → returns w*x + b


class LinearModel:
  def __init__(self, w, b):
    self.w = w
    self.b = b

  def __call__(self, x):
    return self.w * x + self.b

x = 5
w = 2
b = 1
obj_1 = LinearModel(w, b)
print(obj_1(x))
```

```python
# Challenge 11:
#Custom Counter Class

#Implement a class Counter that tracks counts for any added item:

class Counter:
  def __init__(self):
    self.counts = {}

  def add(self, x):
    self.x = x
    if self.x in self.counts:
      self.counts[self.x] += 1
    else:
      self.counts[self.x] = 1


c = Counter()
c.add('apple')
c.add('apple')
c.add('banana')
print(c.counts)  # {'apple': 2, 'banana': 1}

```

```python
# Challenge 12:
#Callable Class for Normalization

class Normalizer:
  def __init__(self, mean, std):
    self.mean = mean
    self.std = std

  def __call__(self, x):
    if(type(x) == int):
      return (x - self.mean) / self.std
    else:
      return ([(i-self.mean)/self.std for i in x])

norm = Normalizer(10, 2)
print(norm(14))         # 2.0
print(norm([8, 10, 14]))  # [-1.0, 0.0, 2.0]
```

```python
# Challenge 13:
# Inheritance Practice — Shape Hierarchy
# Base class Shape with method area()
# Derived classes Circle, Rectangle, and Triangle
# Each should override area() properly.

class Shape:
  def area(self, l, w):
    return l*w

  def area(self):
    return self.l * self.w

class Circle(Shape):
  def area(self, r):
    return (22/7) * (r**2)

class Rectangle(Shape):
  def area(self, l):
    return l**2

class Traingle(Shape):
  def area(self, b, h):
    return (1/2) * b * h


cir = Circle()
print(cir.area(2))
```
