#Missing values

```python
df = df.dropna(subset = ["Price"])
```

```python
df.isnull().sum()
```

```python
df['Horsepower'].fillna(df['Horsepower'].mean(), inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

# Manipulation & Grouping

```python
fuel_tank_affecting_price = df.groupby("Fuel.tank.capacity")["Price"].mean()

print(fuel_tank_affecting_price)
```

```python
brand_price = df.groupby("Manufacturer")["Price"].mean().sort_values(ascending=False)
print(brand_price)
```

# Visualization with Matplotlib

```python
import matplotlib.pyplot as plt

plt.hist(df["Price"], bins = 20)
plt.title("Price Distribution")
plt.xlabel("price")
plt.ylabel("count")
plt.show()

plt.scatter(df['Fuel.tank.capacity'], df['Price'])
plt.title('Fuel.tank.capacity and Price')
plt.xlabel('Fuel.tank.capacity')
plt.ylabel('Price')
plt.show()
```
